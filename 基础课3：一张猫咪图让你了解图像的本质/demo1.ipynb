```python
# 首先导入numpy
import numpy as np
```

```python
# 为了在notebook中显示图片，导入matplotlib库
import matplotlib.pyplot as plt

# 加这行在Notebook显示图像
%matplotlib inline
```

```python
# 再使用一个PIL库，用于读取图像
from PIL import Image
```

```python
# 我在img文件夹下放了一张图片（演示一下）
# 我们用PIL库读取图片，注意路径要正确
img = Image.open('./img/cat.jpg')
```

```python
# 显示图像
img
```

```python
# 可以看到这是一张彩色的猫咪图片
```

```python
# 查看一下变量的类型
type(img)
```

```python
# 可以看到这个不是numpy的数组格式，那numpy还不能处理它
```

```python
# 首先我们需要将它转化为numpy 数组，使用numpy.asarray()函数
img_arr = np.asarray(img)
```

```python
# 查看类型，
type(img_arr)
```

```python
# 发现已经变成了numpy数组，现在我们就可以用numpy来处理它了
```

```python
# 我们查看大小
img_arr.shape
```

```python
# 可以看到这张照片是1880像素宽，1253像素高，3个颜色通道
```

```python
# 再使用matplot的imshow()方法显示Numpy数组形式的图片

plt.imshow(img_arr)
```

```python
# 可以看到横坐标和纵坐标显示了图片的长度是1800多，高度是1200多
```

```python
# 我们继续对这个图片操作，先使用numpy的copy方法复制一份原图
img_arr_copy = img_arr.copy()
```

```python
# 显示一下
plt.imshow(img_arr_copy)
```

```python
# 检查一下大小
img_arr_copy.shape
```

```python
# 首先使用numpy切片，将R,G,B三个颜色通道中的R红色通道显示出来

plt.imshow(img_arr_copy[:,:,0])
```

```python
# 大家会发现这个颜色很奇怪，都是翠绿色，为什么会显示成这样呢？

# 我们打开matplot的官网关于颜色表colormap的说明：
# https://matplotlib.org/stable/gallery/color/colormap_reference.html
# 可以看到默认的颜色：是翠绿色（viridis ）。那这个颜色方便色盲观看的

```

```python
# 我们也可以将cmap颜色设置成火山岩浆样式：magma
plt.imshow(img_arr_copy[:,:,0],cmap='magma')
```

```python
# 我们打印一下红色R通道的数组
img_arr_copy[:,:,0]
```

```python
# 好，我们知道，计算机是分不清到底哪一个通道是红色的，每一个颜色通道其实都是一个灰度图，我们首先将cmap颜色设置为gray灰度
# 看一下
plt.imshow(img_arr_copy[:,:,0],cmap='gray') 
```

```python
# 我们知道，红色通道中的 0呢就是没有红色，代表纯黑色，而越接近255呢，就代表越红，255就纯红色
# 那看这个灰度图，颜色越浅，表示这里越红 
# 我们可以看一下红色通道的灰度图上颜色最浅的就是这个吊坠（鼠标指示）
# 那回到原来彩色图片，可以看到这个吊坠确实是最红的
```

```python
# 类似的，我们将绿色通道也显示为灰度模式
plt.imshow(img_arr_copy[:,:,1],cmap='gray')
```

```python
# 那 0呢代表没有绿色或纯黑色，255呢就代表纯绿色
# 可以看到，灰度图上颜色越浅，表示这里越绿
```

```python
# 再看一下蓝色通道

plt.imshow(img_arr_copy[:,:,2],cmap='gray')
```

```python
# 0：没有蓝色或纯黑色，255代表纯蓝色
# 灰度图颜色越浅，表示这里越蓝 ，可以看到这里相比前面红色、绿色的灰度图，这个花瓶是颜色比较浅的，代表颜色接近蓝色
```

```python
# 当然，我们可以将某个颜色通道颜色全部设为0，我们看一下效果，
# 我这里把绿色通道全部变为0
img_arr_copy[:,:,1] = 0
```

```python
# 显示一下
plt.imshow(img_arr_copy)
```

```python
# 那会发现画面颜色特别紫，这是因为只剩红色、蓝色通道，而（红+蓝就是紫色）
```

```python
# 我们再将蓝色通道全部变为0
img_arr_copy[:,:,2] = 0
```

```python
# 显示一下
plt.imshow(img_arr_copy)
```

```python
# 那只剩红色通道了，所以画面特别红，
# 那这个红色图片和前面的红色灰度图之所不同，是因为现在是我们三个通道一起合成在看
```

```python
# 可以查看大小shape，会发现大小仍然不变
img_arr_copy.shape
```

```python
# 而单独看一个通道的时候，大小会变化
img_arr_copy[:,:,0].shape
```
